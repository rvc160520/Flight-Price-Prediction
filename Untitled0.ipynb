```python
#function for converting time of format "02h 05m" to minutes
def convert_to_minutes(time_taken):

    t = time_taken[:-1].split('h ')
    total_minutes = int(t[0]) * 60 + int(t[1])
    return total_minutes
df['Duration_min'] = df['Duration'].apply(convert_to_minutes)
```

```python
#function for converting time of format 00:00 to minutes
def convert_to_minutes(time_taken):

    t = time_taken.split('\n')
    t_split=t[0].split(':')
    total_minutes = int(t_split[0]) * 60 + int(t_split[1])
    return total_minutes
```

```python
df['dept_time'] = df['Departure Time'].apply(convert_to_minutes)
df['arr_time'] = df['Arrival Time'].apply(convert_to_minutes)
```

```python
def calculate_performance(y_train,y_pred):
    mse = mean_squared_error(y_train, y_pred)
    mae = mean_absolute_error(y_train, y_pred)
    r2 = r2_score(y_train, y_pred)
    rmse = np.sqrt(mse)

    return [rmse, mse, mae, r2]
score_names = ['RMSEs', 'MSEs', 'MAEs', 'R2s']
```

```python
RMSEs = []
R2s   = []
MSEs  = []
MAEs  = []
l =["RMSE","MSE","MAE","R2",]
for idx in range(len(X_train_list)):
  dt_idx = DTR()
  dt_idx.fit(X_train_list[idx],Y_train_list[idx])
  dt_pred_idx = dt_idx.predict(X_test_list[idx])
  dt_scores_idx = calculate_performance(Y_test_list[idx],dt_pred_idx)
  #print(split_data_name[idx])
  RMSEs.append(dt_scores_idx[0])
  MSEs.append(dt_scores_idx[1])
  MAEs.append(dt_scores_idx[2])
  R2s.append(dt_scores_idx[3])
  # for name, score in zip(score_names, dt_scores_idx):
  #   print("{:10} : {:.5f}".format(name, score))

evalution_df = pd.DataFrame({
  'feature': split_data_name,
  'RMSE': RMSEs,
  'MSE' : MSEs,
  'MAE' : MAEs,
  'R2s' : R2s,
}).sort_values('R2s', ascending=False)
evalution_df.head(10)
```
